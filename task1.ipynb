```python
import numpy as np
import pandas as pd
```

```python
data = np.load("./QIS_EXAM_200Events.npz",allow_pickle=True,fix_imports = True)
```

```python
train = data['training_input'][()] # or use .item()
test = data['test_input'].item()
```

#### Loading and preparing the dataset

```python
x_train=pd.concat([pd.DataFrame(train['0']),pd.DataFrame(train['1'])],ignore_index=True)

x_test=pd.concat([pd.DataFrame(test['0']),pd.DataFrame(test['1'])],ignore_index=True)
```

```python
y_train =pd.concat([pd.DataFrame(np.zeros(50),columns=['out']),pd.DataFrame(np.ones(50),columns=['out'])],axis = 0,ignore_index=True)

y_test =pd.concat([pd.DataFrame(np.zeros(50),columns=['out']),pd.DataFrame(np.ones(50),columns=['out'])],axis = 0,ignore_index=True)
```

```python
x_train = x_train*100
x_test = x_test * 100
```

#### Training and testing a SVM for classification

```python
from sklearn.svm import SVC,NuSVC,LinearSVC

model1 = SVC(degree = 3,kernel='poly',verbose = 1)
model1.fit(x_train,y_train)

# model2 = NuSVC()
# model2.fit(x_train.iloc[:,:],y_train.iloc[:,:])
```

```python
from sklearn.metrics import accuracy_score

y_pred1 = model1.predict(x_test)
print(accuracy_score(y_pred1,y_test))

# ypred2 = model2.predict(x_test.iloc[:,:])
# print(accuracy_score(ypred2,y_test))
```

#### Fine tuning the model

```python
Cs = [0.001,0.01,0.1,1,10]
gammas = ['scale','auto']
degrees = [2,3,4]
kernels = ['poly','rbf']
param_grid = {'C':Cs,'gamma':gammas,'kernel':kernels,'degree':degrees}
```

```python
from sklearn.model_selection import GridSearchCV

g_s = GridSearchCV(SVC(), param_grid)
```

```python
g_s.fit(x_train,y_train)
```

```python
g_s.best_params_
```

```python
y_pred2 = g_s.predict(x_test)
print(accuracy_score(y_pred2,y_test))
```

Fine tuning overfits the train set thus giving less accuracy in the test set.
